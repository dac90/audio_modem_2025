```python
import sounddevice as sd
import scipy.io.wavfile as wav
import numpy as np
import os
import scipy.io.wavfile
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
import numpy as np
import os
import scipy.io.wavfile
import matplotlib.pyplot as plt
import scipy.signal
# Parameters for recording
DURATION = 15  # Duration of the recording in seconds
FS = 48000    # Sampling frequency (samples per second)

# Function to record audio
def record_audio(duration, fs):
    print("Recording...")
    audio = sd.rec(int((duration) * fs), samplerate=fs, channels=2, dtype='int16')
    sd.wait()  # Wait until the recording is finished
    print("Recording finished.")
    return audio

# Function to save audio to a WAV file
def save_audio(filename, audio, fs):
    wav.write(filename, fs, audio)
    print(f"Audio saved to {filename}")

# Main program
if __name__ == "__main__":
    output_file = "testgroup4.wav"  # Name of the output WAV file
    recorded_audio = record_audio(DURATION, FS)
    save_audio(output_file, recorded_audio, FS)
    print("Current working directory:", os.getcwd())


_,  sig = scipy.io.wavfile.read("testgroup4.wav")

fig, axs = plt.subplots(2, figsize=(8, 10))  # Create two subplots
axs: list[plt.Axes]  # axs is a list of Axes objects

if sig.ndim == 2:
    sig = sig[:, 0]  # Select the first channel (left channel)
    # Alternatively, convert to mono by averaging the channels:
    # sig = np.mean(sig, axis=1)

# Plot the spectrogram on the first axis
f, t_spec, Sxx = scipy.signal.spectrogram(sig, FS)
pcm = axs[0].pcolormesh(t_spec, f, 10 * np.log10(Sxx), shading='gouraud')
axs[0].set_ylabel('Frequency [Hz]')
axs[0].set_xlabel('Time [sec]')
axs[0].set_title('Spectrogram')
cbar = fig.colorbar(pcm, ax=axs[0], label='Power/Frequency (dB/Hz)')

# Optionally, you can use the second axis for another plot (e.g., waveform)
axs[1].plot(sig)
axs[1].set_title('Waveform')
axs[1].set_xlabel('Time [samples]')
axs[1].set_ylabel('Amplitude')

# Show the plots
plt.tight_layout()
plt.show()


```
